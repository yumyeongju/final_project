```python
def preprocess_yearly_data(df):
    """
    연도별 데이터에서 0행을 열 이름으로 설정하고, 그 아래 데이터만 남기는 함수.
    """
    df.columns = df.iloc[0]
    df = df.drop(index=0).reset_index(drop=True)
    return df
```

```python
# 삶의 만족도
df_satisfaction_2019 = preprocess_yearly_data(df_satisfaction_2019)
df_satisfaction_2020 = preprocess_yearly_data(df_satisfaction_2020)
df_satisfaction_2021 = preprocess_yearly_data(df_satisfaction_2021)
df_satisfaction_2022 = preprocess_yearly_data(df_satisfaction_2022)
df_satisfaction_2023 = preprocess_yearly_data(df_satisfaction_2023)
df_satisfaction_2024 = preprocess_yearly_data(df_satisfaction_2024)

# 문화생활 참여
df_culture_2019 = preprocess_yearly_data(df_culture_2019)
df_culture_2021 = preprocess_yearly_data(df_culture_2021)
df_culture_2023 = preprocess_yearly_data(df_culture_2023)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 시각화 스타일 설정
sns.set_style("whitegrid")
plt.rcParams['font.family'] = 'NanumGothic'

# 연도별 데이터
years = ['2019', '2021', '2023']
culture_activities = ['- 음악·연주회·콘서트 (%)', '- 영화 (%)', '- 스포츠 경기 (%)']
satisfaction_levels = ['매우 만족', '약간 만족', '보통', '약간 불만족', '매우 불만족']
genders = ['남자', '여자']
colors_gender = {'남자': 'skyblue', '여자': 'salmon'}
colors_satisfaction = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd'] # 각 만족도 수준별 색상

# 그래프 그리기 (성별 분리, 만족도별 색상 구분)
for gender in genders:
    for activity in culture_activities:
        plt.figure(figsize=(14, 8))

        # 문화 활동 참여율
        participation_rates = [
            df_culture_2019[df_culture_2019['특성별(2)'] == gender][activity].values[0],
            df_culture_2021[df_culture_2021['특성별(2)'] == gender][activity].values[0],
            df_culture_2023[df_culture_2023['특성별(2)'] == gender][activity].values[0]
        ]
        plt.plot(years, participation_rates, marker='o', linestyle='-', label=f'{activity} 참여율', color=colors_gender[gender])
        plt.ylabel('참여율 (%)', color=colors_gender[gender], fontsize=12)
        plt.tick_params(axis='y', labelcolor=colors_gender[gender])
        plt.legend(loc='upper left')
        plt.twinx()

        # 삶의 만족도
        satisfaction_dfs = {
            '2019': df_satisfaction_2019,
            '2021': df_satisfaction_2021,
            '2023': df_satisfaction_2023
        }
        satisfaction_columns = {
            '2019': ['매우 만족', '약간 만족', '보통', '약간 불만족', '매우 불만족'],
            '2021': ['매우 만족', '약간 만족', '보통', '약간 불만족', '매우 불만족'],
            '2023': ['매우 만족', '약간 만족', '보통', '약간 불만족', '매우 불만족']
        }

        for i, level in enumerate(satisfaction_levels):
            satisfaction_rates = []
            for year in years:
                df_sat = satisfaction_dfs[year]
                sat_col = satisfaction_columns[year][i]
                rate = df_sat[df_sat['특성별(2)'] == gender][sat_col].values[0]
                satisfaction_rates.append(rate)
            plt.plot(years, satisfaction_rates, marker='x', linestyle='--', label=f'삶의 만족도 ({level})', color=colors_satisfaction[i])

        plt.title(f'코로나19 이전 및 이후 {activity} 참여율 변화와 삶의 만족도 추이 비교 ({gender})', fontsize=14)
        plt.xlabel('연도', fontsize=12)
        plt.ylabel('삶의 만족도 (%)', color='black', fontsize=12)
        plt.tick_params(axis='y', labelcolor='black')
        plt.legend(loc='upper right')
        plt.grid(True)
        plt.tight_layout()
        plt.show()
```
